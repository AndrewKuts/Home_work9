```python
import cv2
import numpy as np
import math
from matplotlib import pyplot as plt
import dlib
plt.rcParams['figure.figsize'] = [15, 10]
```

```python
img = cv2.imread('eu_summit.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
detector = dlib.get_frontal_face_detector()
rects = detector(gray, 1)
```

```python
print('Number of detected faces:', len(rects))
print(rects)
print(rects[0].left)
def rect_to_bb(rect):
    # Dlib rect --> OpenCV rect
    x = rect.left()
    y = rect.top()
    w = rect.right() - x
    h = rect.bottom() - y
    return (x, y, w, h)
```

```python
result_dlib = np.copy(img)
faces_dlib_img = []
import random
for rect in rects:    
    x, y, w, h = rect_to_bb(rect)
    print(x, y, w, h)
    r = random.randint(0,255)
    g = random.randint(0,255)
    b = random.randint(0,255)
    cv2.rectangle(result_dlib, (x, y), (x+w, y+h), (r, r, b), 2)
    faces_dlib_img.append(img[y:y+h, x:x+w, :])
plt.imshow(result_dlib), plt.title('dlib')
```
